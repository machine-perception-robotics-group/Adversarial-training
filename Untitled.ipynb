```python
import os
import torch
import torch.nn as nn
from tqdm import tqdm
from utils import *

model_path = './checkpoint/kernel_trick/best_model'
os.environ['CUDA_VISIBLE_DEVICES'] = '0,1'
_, test_dataloader = get_dataloader(dataset='mnist', batch_size=100, image_size=28)
model = nn.DataParallel(get_network(model_type='rbf_cnn', num_classes=10).cuda())
state_dict = torch.load(model_path)['state_dict']
model.load_state_dict(state_dict)
model.eval()

epsilon = 8/255
alpha = 2/255
total_acc_nat = 0
total_acc_rob = 0
xent = nn.CrossEntropyLoss()
total = 0
for ind, samples in enumerate(tqdm(test_dataloader)):
    inputs, targets = samples[0].cuda(), samples[1].cuda()
    noise = torch.FloatTensor(inputs.shape).uniform_(-1, 1).cuda()
    x = torch.clamp(inputs+noise, min=0, max=1)
    
    for _ in range(10):
        x.requires_grad_()
        logits, _ = model(x)
        loss = xent(logits, targets)
        loss.backward()
        grads = x.grad.data
        x = x.data.detach() + alpha*torch.sign(grads).detach()
        x = torch.min(torch.max(x, inputs - epsilon), inputs + epsilon)
        x = torch.clamp(x, min=0, max=1)
    
    with torch.no_grad():
        logits_nat, _ = model(inputs)
        logits_adv, _ = model(x)
        
    total += inputs.size(0)
    total_acc_nat += logits_nat.softmax(dim=1).argmax(dim=1).eq(targets).sum().item()
    total_acc_rob += logits_adv.softmax(dim=1).argmax(dim=1).eq(targets).sum().item()
    
print('total: %d' % (total))
print('Acc (nat): %.4f' % (total_acc_nat/len(test_dataloader.dataset)))
print('Acc (rob): %.4f' % (total_acc_rob/len(test_dataloader.dataset)))
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, mutual_info_score, homogeneity_score, completeness_score
from bhtsne import tsne
%matplotlib inline
import matplotlib.pyplot as plt

cnt = 0
Ls = torch.zeros(len(test_dataloader.dataset))
Ps = torch.zeros(len(test_dataloader.dataset))
Fs = torch.zeros(len(test_dataloader.dataset), 32)
for samples in test_dataloader:
    inputs, targets = samples[0].cuda(), samples[1].cuda()
    batch = inputs.size(0)
    
    with torch.no_grad():
        logits, features = model(inputs)
        
    Ps[cnt:cnt+batch] = logits.softmax(dim=1).argmax(dim=1).data.cpu()
    Ls[cnt:cnt+batch] = targets.data.cpu()
    Fs[cnt:cnt+batch] = features.data.cpu().view(batch, -1)
    cnt += batch
    
X = tsne(Fs.to(torch.float64).numpy(), dimensions=2, perplexity=30, rand_seed=1729)
xcoords, ycoords = X[:, 0], X[:, 1]
color_list = plt.rcParams['axes.prop_cycle'].by_key()['color']
colorlist = [color_list[int(idx.item())] for idx in Ls]

plt.scatter(xcoords, ycoords, color=colorlist, s=0.4)
plt.show()
```

```python
sil_score = silhouette_score(Fs.numpy(), Ls.numpy())
cal_score = calinski_harabasz_score(Fs.numpy(), Ls.numpy())
mi_score = mutual_info_score(Ls, Ps)
homo_score = homogeneity_score(Ls, Ps)
comp_score = completeness_score(Ls, Ps)
print('Sil. score: %.6f' % sil_score)
print('Cal. score: %.6f' % cal_score)
print('MI score: %.6f' % mi_score)
print('Homo score: %.6f' % homo_score)
print('Comp score: %.6f' % comp_score)
```
